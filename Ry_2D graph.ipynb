```python
#imports
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdul = fits.open('Ry15_WCS2.fits')
Ry15 = hdul[0].data

hdul = fits.open('Ry42_OS.fits')
Ry42 = hdul[0].data
#dtnocor = nocorrection[0].astype(int)
```

```python
#function that prints 2dimentional array dimensions
def dim(array):
    len1 = str(len(array))
    len2 = str(len(array[0]))
    prt = "Dimension: " + len1 + " × " + len2
    return (prt)
```

```python
print(dim(Ry15))
print(dim(Ry15[1]))

print(dim(Ry42))
```

```python
f = Ry15[1]
plt.imshow(f, interpolation="nearest", origin="upper")
plt.colorbar()
plt.show()
```

```python
f = Ry15[1][3700:4200][:,2200:2700]
plt.imshow(f, interpolation="nearest", origin="upper")
plt.colorbar()
plt.show()
```
