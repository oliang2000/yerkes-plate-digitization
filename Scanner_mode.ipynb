```python
#function that prints 2dimentional array dimensions
def dim(array):
    len1 = str(len(array))
    len2 = str(len(array[0]))
    prt = "Dimension: " + len1 + " × " + len2
    return (prt)
```

```python
#imports
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import fits data
hdul = fits.open('scannertest/ngc2392_1_001.fits')
nc1 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_1_002.fits')
nc2 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_1_003.fits')
nc3 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_1_004.fits')
nc4 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_1_005.fits')
nc5 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_2_006.fits')
ccauto1 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_2_007.fits')
ccauto2 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_2_008.fits')
ccauto3 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_2_009.fits')
ccauto4 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_2_010.fits')
ccauto5 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_3_011.fits')
ccfull1 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_3_012.fits')
ccfull2 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_3_013.fits')
ccfull3 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_3_014.fits')
ccfull4 = hdul[0].data
hdul = fits.open('scannertest/ngc2392_3_015.fits')
ccfull5 = hdul[0].data
```

```python
#comparing a cut across the planetary nebula (NGC2392)

#no correction
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
x1 = 2500
x2 = 3000
y = 2535
axes.plot(list(nc1[0][y][x1:x2]))
axes.plot(list(nc2[0][y][x1:x2]))
axes.plot(list(nc3[0][y][x1:x2]))
axes.plot(list(nc4[0][y][x1:x2]))
axes.plot(list(nc5[0][y][x1:x2]))
plt.show()

#correction, auto histogram
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(list(ccfull1[0][y][x1:x2]))
axes.plot(list(ccfull2[0][y][x1:x2]))
axes.plot(list(ccfull3[0][y][x1:x2]))
axes.plot(list(ccfull4[0][y][x1:x2]))
axes.plot(list(ccfull5[0][y][x1:x2]))
plt.show()

#correction, auto histogram but output 0-255
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(list(ccauto1[0][y][x1:x2]))
axes.plot(list(ccauto2[0][y][x1:x2]))
axes.plot(list(ccauto3[0][y][x1:x2]))
axes.plot(list(ccauto4[0][y][x1:x2]))
axes.plot(list(ccauto5[0][y][x1:x2]))
plt.show()
```

- The color corrections saturate the star curves, causing a loss of information. There seems to be a larger S/N ratio for the corrected curves, but that is just how the histogram correction decrease the amplitude of low numbers to increase contrast.

```python
#comparing a cut across noise

#no correction
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
x1 = 3000
x2 = 3200
y = 2535
axes.plot(list(nc1[0][y][x1:x2]))
axes.plot(list(nc2[0][y][x1:x2]))
axes.plot(list(nc3[0][y][x1:x2]))
axes.plot(list(nc4[0][y][x1:x2]))
axes.plot(list(nc5[0][y][x1:x2]))
plt.show()

#correction, auto histogram
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(list(ccfull1[0][y][x1:x2]))
axes.plot(list(ccfull2[0][y][x1:x2]))
axes.plot(list(ccfull3[0][y][x1:x2]))
axes.plot(list(ccfull4[0][y][x1:x2]))
axes.plot(list(ccfull5[0][y][x1:x2]))
plt.show()

#correction, auto histogram but output 0-255
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(list(ccauto1[0][y][x1:x2]))
axes.plot(list(ccauto2[0][y][x1:x2]))
axes.plot(list(ccauto3[0][y][x1:x2]))
axes.plot(list(ccauto4[0][y][x1:x2]))
axes.plot(list(ccauto5[0][y][x1:x2]))
plt.show()
```

- With color correction auto histogram, the higher numbers are saturated.
