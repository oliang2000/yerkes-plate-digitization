```python
#imports
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdul = fits.open('scannertest/ry65_16gray_1400_pf_nocorrection.fits')
nocorrection = hdul[0].data

hdul = fits.open('scannertest/ry65_16gray_1400_pf_colorcontro22.fits')
colorcontro22 = hdul[0].data
#dtnocor = nocorrection[0].astype(int)
```

```python
#function that prints 2dimentional array dimensions
def dim(array):
    len1 = str(len(array))
    len2 = str(len(array[0]))
    prt = "Dimension: " + len1 + " × " + len2
    return (prt)
```

```python
print(dim(nocorrection))
```

```python
fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
x1 = 1856
x2 = 2321
y = 2005
axes.plot(list(nocorrection[y][x1:x2]), alpha=1, color = 'black')
plt.show()

fig = plt.figure(figsize=(20,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
x1 = 1856
x2 = 2321
y = 2005
axes.plot(list(colorcontro22[y][x1:x2]), alpha=1, color = 'red')
plt.show()
```
